```python
import pandas as pd
import re
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

```python
stop_words = stopwords.words('english')
len(stop_words)
```

```python
sms = pd.read_csv('c:/data/sms_spam1.csv')
sms
```

```python
docs = [s for s in sms.text]
docs
```

```python
cv = CountVectorizer(binary=False,token_pattern="\\b[a-z]+\\b",
                    stop_words=stop_words,lowercase=True,ngram_range=(1,1))
bow = cv.fit_transform(docs).toarray()
bow
```

```python
cv.get_feature_names_out()
```

```python
len(cv.get_stop_words()),len(stop_words)
```

```python
cv.get_stop_words().difference(stop_words)
```

```python
df_words = pd.DataFrame(bow,columns=cv.get_feature_names_out())
df_words
```
