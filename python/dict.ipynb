```python
nums =[2,4,1,2,4,3,2,1]

freq = {}

for n in nums:
    freq[n]= freq.get(n,0)+1

print(freq)
```

```python
from collections import Counter

nums = [2,4,1,2,4,3,2,1]
print(Counter(nums))
```

```python
l = [6,8,6,8,9,1,2,9,3,5,7,11,1,66,2]

m = {}
p = []
for n in l:
    if n in m:
        m[n] = m[n] + 1
    else:
        m[n] = 1
        p.append(n)

print(p)
```
