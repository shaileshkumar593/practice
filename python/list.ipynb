```python
def subarray_sum_with_elements(nums, k):
    count = 0
    prefix = 0

    # prefix_sum -> list of indices
    mp = {0: [-1]}

    result = []

    for i, num in enumerate(nums):
        prefix += num

        target = prefix - k

        if target in mp:
            for start_idx in mp[target]:
                count += 1
                result.append(nums[start_idx + 1:i + 1])

        mp.setdefault(prefix, []).append(i)

    return count, result


count, subarrays = subarray_sum_with_elements([2,1, 4, 7, 8], 7)

print("Count =", count)
print("Subarrays =", subarrays)
```

```python
#
def longest_subarray_sum_k(nums, k):
    prefix = 0
    max_len = 0
    start = end = -1 
    # Prefix sum 0 occurred before the array started,at index -1.

    mp = {0: -1}

    for i, num in enumerate(nums):
        prefix += num

        if prefix - k in mp:
            length = i - mp[prefix - k]

            if length > max_len:
                max_len = length
                start = mp[prefix - k] + 1
                end = i

        if prefix not in mp:
            mp[prefix] = i

    return max_len, nums[start:end+1]


length, arr = longest_subarray_sum_k(
    [1,2,3,1,1,1,1],
    6
)

print(length)
print(arr)
```

```python
# sub array exist 

def subarray_exists(nums, k):
    prefix = 0
    seen = {0}

    for num in nums:
        prefix += num

        if prefix - k in seen:
            return True

        seen.add(prefix)

    return False


print(subarray_exists([3, 2, 4, 7, 8], 7))
```

```python
def max_area(height):
    left = 0
    right = len(height) - 1

    ans = 0

    while left < right:
        width = right - left
        area = min(height[left], height[right]) * width
        ans = max(ans, area)

        if height[left] < height[right]:
            left += 1
        else:
            right -=1

    return ans

print(max_area([1,8,6,2,5,4,8,3,7]))
```

```python
def length_of_longest_substring(s):
    seen = set()

    left = 0
    ans = 0

    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1

        seen.add(s[right])
        ans = max(ans,right - left + 1)

    return ans


print(length_of_longest_substring("abcabcbb"))
```

```python
# nge

def next_greater(nums):
    n = len(nums)

    result = [-1]*n
    stack = []

    for i in range(n):
        while stack and nums[i] > nums[stack[-1]]:
            idx = stack.pop()
            result[idx] = nums[i]
        stack.append(i)

    return result

print(next_greater([5,4,3,6,9]))
```

```python
def first_occurence(nums, target):
    left = 0
    right= len(nums)-1
    answer = -1

    while left <= right:
        mid = (left + right)//2

        if nums[mid] == target:
            answer = mid
            right = mid -1 # to get leftmost target point
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid -1

    return answer

print(first_occurence([1,2,2,2,3], 2))
```

```python
# merge interval 

def merge(intervals):
    intervals.sort()

    result = [intervals[0]]
    print(result)

    for start ,end in intervals[1:]:
        last_end = result[-1][1]

        if start <= last_end:
            result[-1][1] = max(last_end, end)

        else:
            result.append([start, end])

    return result


print(merge([[1,3],[2,6],[8,10],[15,18]]))
```

```python
def remove_duplicates(nums):

    if not nums:
        return 0

    slow = 1

    for fast in range(1, len(nums)):

        if nums[fast] != nums[fast - 1]:
            nums[slow] = nums[fast]
            slow += 1

    return slow


nums = [1,1,2,2,3]

k = remove_duplicates(nums)

print(nums[:k])
```

```python
# kth largest 

import heapq

def kth_largest(nums,k):
    heap = []
    
    for num in nums:
        heapq.heappush(heap, num)

        if len(heap) > k:
            heapq.heappop(heap)

    return heap[0]


print(kth_largest([3,2,1,5,6,4], 2))
```

```python
#sliding window 
# works for positive element 

def subarray_sum(nums, target):
    left = 0
    curr_sum = 0

    for right in range(len(nums)):
        curr_sum += nums[right]

        while curr_sum > target:
            curr_sum -= nums[left]
            left += 1

        if curr_sum == target:
            return nums[left:right+1]

    return []


print(subarray_sum([1,2,4,3,7,5], 10))
```

```python
def all_subarrays_sum(nums, target):
    left = 0
    curr_sum = 0
    result = []

    for right in range(len(nums)):
        curr_sum += nums[right]

        while curr_sum > target and left <= right:
            curr_sum -= nums[left]
            left += 1

        if curr_sum == target:
            result.append(nums[left:right+1])

    return result


print(all_subarrays_sum([1,2,4,3,7,5], 10))
```

```python
# all possible sum 

def combination_Sum(nums,target):
    result = []

    def BackTracking(start,path, remaining):
        if remaining == 0:
            result.append(path[:])

        if remaining < 0:
            return
        
        for i in range(start,len(nums)):
            path.append(nums[i])
            BackTracking(i, path, remaining-nums[i])

            path.pop()

    BackTracking(0,[],target)

    return result


nums = [2,3,6,7]
target = 7

print(combination_Sum(nums, target))
```

```python
import time
from collections import defaultdict, deque


class IPRateLimiter:

    def __init__(self, limit, window_seconds):
        self.limit = limit
        self.window = window_seconds
        self.requests = defaultdict(deque)

    def allow(self, ip):

        now = time.time()

        q = self.requests[ip]

        # Remove expired timestamps
        while q and q[0] <= now - self.window:
            q.popleft()

        # Check limit
        if len(q) >= self.limit:
            return False

        # Store current request
        q.append(now)

        return True


limiter = IPRateLimiter(
    limit=5,
    window_seconds=60
)

ip = "192.168.1.10"

for i in range(7):
    print(f"Request {i+1}: {limiter.allow(ip)}")
```

```python
from fastapi import FastAPI, Request
from fastapi.responses import JSONResponse

app = FastAPI()

limiter = IPRateLimiter(
    limit=5,
    window_seconds=60
)


@app.middleware("http")
async def rate_limit(request: Request, call_next):

    ip = request.client.host

    if not limiter.allow(ip):
        return JSONResponse(
            status_code=429,
            content={
                "message": "Too Many Requests"
            }
        )

    return await call_next(request)


@app.get("/")
async def home():
    return {"message": "success"}
```

```python
from concurrent.futures import ThreadPoolExecutor
import threading

cache = {}
lock = threading.Lock()

def square(n):

    result = n * n

    with lock:
        cache[n] = result

    return result


with ThreadPoolExecutor(max_workers=5) as executor:

    futures = [
        executor.submit(square, i)
        for i in range(10)
    ]

    for future in futures:
        print(future.result())

print(cache)
```
