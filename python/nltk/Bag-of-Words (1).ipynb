```python
import pandas as pd
import re
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
len(stop_words)
```

```python
sms = pd.read_csv('c:/data/sms_spam1.csv')
sms
```

```python
docs = []
for s in sms.text:
    words = re.findall(r'\b[a-z]+\b',s.lower())
    clean_words = [w for w in words if w not in stop_words]
    docs.append(" ".join(clean_words))
docs
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(binary=True)
bow = cv.fit_transform(docs).toarray()
bow
```

```python
cv.get_feature_names_out()
```

```python
df_words = pd.DataFrame(bow,columns=cv.get_feature_names_out())
df_words
```
