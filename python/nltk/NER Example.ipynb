```python
import nltk
from nltk import word_tokenize,pos_tag,ne_chunk
```

```python
nltk.download('words')
```

```python
text = "Bill Gates founded Microsoft in Albuquerque, New Mexico on 01-jan-2000 with a value of USD 200."
```

```python
words = word_tokenize(text)
words
```

```python
pos = pos_tag(words)
pos
```

```python
ner = ne_chunk(pos)
print(ner)
```

```python
import spacy

# Load the English model
nlp = spacy.load("en_core_web_sm")

# Process your text
text = "kohli likes to eat biryani in india which costs  $2000 at google"
doc = nlp(text)

# Print detected entities
for ent in doc.ents:
    print(f"Entity: {ent.text} | Label: {ent.label_} ({spacy.explain(ent.label_)})")
```

```python
from spacy import displacy

# Visualize the entities in a browser or notebook
displacy.render(doc, style="ent")
```
