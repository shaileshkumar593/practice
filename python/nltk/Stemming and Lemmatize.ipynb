```python
words = ['walking','walks','walked','activated','beautiful','better','best','mens','yours','his','mice','good']
```

```python
port_stem = PorterStemmer()
snow_stem = SnowballStemmer(language='english',ignore_stopwords=True)
lemma = WordNetLemmatizer()
```

```python
[port_stem.stem(w) for w in words]
```
