```python
import pandas as pd 
import re
from wordcloud import WordCloud
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
```

```python
stop_words = stopwords.words('english')
```

```python
sms = pd.read_csv('c:/data/sms_spam.csv')
sms
```

```python
text  =[]
for s in sms.text:
    words = re.findall(r'\b[a-z]+\w\b',s.lower())
    text.extend([w for w in words if w not in stop_words])
```

```python
plt.figure(figsize=[12,6])
wc = WordCloud(width=1200,height=600,max_words=200,repeat=False).generate_from_text(" ".join(text))
plt.imshow(wc)
plt.show()
```
